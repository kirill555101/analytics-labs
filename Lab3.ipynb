## Задание 1

Создайте гистограмму, которая будет отражать показатели посещаемости студентов вашей группы на лабораторных работах за последние 2 недели.

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
visits = [1, 1, 2, 2, 2, 2, 3, 4, 4, 4]

plt.hist(visits)
plt.xlabel('Количество посещений')
plt.ylabel('Число студентов')
plt.title('Гистограмма')
plt.show()
```

## Задание 2

Создайте Bar Chart , которая будет отражать показатели посещаемости студентов вашей группы на лабораторных работах за последние 2 недели, добавьте в вывод что наиболее удобно для работы и почему.

```python
days = ['Пн', 'Вт', 'Ср', 'Чт', 'Пт']

plt.bar(days * 2, visits)
plt.xlabel('День')
plt.ylabel('Количество посещений')
plt.title('Bar Chart')
plt.show()
```

## Задание 3

Нарисуйте 3 уровня лестницы с помощью Stacked Area Chart различными цветами, добавьте лейблы и название графика.

```python
level_1 = [1, 2, 2, 4, 5]
level_2 = [1, 3, 5, 5, 9]
level_3 = [1, 7, 7, 10, 13]

plt.stackplot(days, level_1, level_2, level_3)
plt.xlabel('День')
plt.ylabel('Высота уровня')
plt.title('Stacked Area Chart')
plt.show()
```

## Дополнительное задание

Количество диабетиков для мужчин и женщин, снизу - возраст. Диаграмма - Stacked Area Chart

```python
df = pd.read_csv('data/diabetes_dataset.csv')
df.head()
```

```python
elements_count = 75

ages = df['age'].sort_values().unique().take(range(elements_count))
male_count = df[(df.gender == 'Male') & (df.diabetes == 1)]\
    .groupby('age').size()\
    .head(elements_count).values
female_count = df[(df.gender == 'Female') & (df.diabetes == 1)]\
    .groupby('age').size()\
    .head(elements_count).values

plt.stackplot(ages, male_count, female_count - male_count, labels=['Мужчины', 'Женщины'])
plt.legend()
plt.xlabel('Возраст')
plt.ylabel('Количество диабетиков')
plt.title('Stacked Area Chart')
plt.show()
```
